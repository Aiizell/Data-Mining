```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("vlr.csv")
data.head()
```

```python
data.info()
```

```python
#mencari data unique
data["Winner"].unique()
```

```python
win_team = 'Paper Rex'
data[data['Winner']==win_team]
```

```python
team_name = data[data.FK_all>0]['Team'].unique()
team_name.sort()
team_name
```

```python
team_name_case=[]
for i in team_name:
  team_name_case.append(data[data.FK_all>0][data['Team']==i]
                                [data['Winner']==win_team].FK_all.sum())
```

```python
print(set(zip(team_name, team_name_case)))
```

```python
nan_indices = []
for i in range(len(team_name)):
  if type(team_name[i])==float:
    nan_indices.append(i)
```

```python
team_name = list(team_name)
team_name_case = list(team_name_case)
```

```python
for i in nan_indices:
  team_name.pop(i)
  team_name_case.pop(i)

team_name
```

```python
plt.barh(team_name, team_name_case)
plt.show()
```

```python
plt.figure(figsize=(10,20))
plt.barh(team_name, team_name_case)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(team_name_case)
plt.legend(team_name, loc='best')
plt.show()
```
